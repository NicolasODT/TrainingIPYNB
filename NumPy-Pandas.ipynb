```python
df2.iloc[2:5] # on affuche les lignes 2 à 5 du DataFrame mais .iloc[] est une meilleure pratique que df2[2:5]
```

```python
df2[['int', 'date', 'email2']] # on affiche les colonnes int, date et email2 du DataFrame
```

```python
thirdRow = [True if i == 500 else False for i in range(len(df2))] # on crée une liste de booléens avec True si i == 500 sinon False
#len() = longueur de la liste si 1000 éléments len() = 1000 
df2[thirdRow]
```

```python
condition = df2['int'] > 990 # on crée une liste de booléens avec True si la valeur de la colonne int est supérieur à 900 sinon False
df2[condition]
```

```python
df2[(df2['float'] > 500) & (df2['int'] > 990)] # on affiche les lignes du DataFrame où la valeur de la colonne float est supérieur à 500 et la valeur de la colonne int est supérieur à 990
```

```python
df2[(df2['float'] > 994) | (df2['int'] > 990)] # on affiche les lignes du DataFrame où la valeur de la colonne float est supérieur à 994 ou la valeur de la colonne int est supérieur à 990
```
